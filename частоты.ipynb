```python
import pandas as pd
df1 = pd.read_csv("princess_corpus.csv")
df = df1.copy()
df.head()
```

```python
import nltk
```

```python
from pymorphy2 import MorphAnalyzer
from nltk import wordpunct_tokenize
morph = MorphAnalyzer()
```

```python
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
```

```python
def lemmatize(x):
    if type(x) != str:
        return ""
    text = wordpunct_tokenize(x)
    result = []
    for word in text:
        if word.isalpha():
            nf = morph.parse(word)[0].normal_form
            if nf not in stops:
                result.append(nf)
    return " ".join(result)
```

```python
from pymystem3 import Mystem
m = Mystem()
```

```python
def parse(x):
    result = []
    ana = nltk.pos_tag(x.split())
    for tagged_token in ana:
        result.append(tagged_token[1])
    return ' '.join(result)
```

```python
def length(x):
    return len(x)
```

```python
df['Length'] = (df['Text']).apply(length)
df
```

```python
df['Text'] = df['Text'].apply(lemmatize)
df['POS'] = df['Text'].apply(parse)
df
```

```python
mean = df['Length'].mean()
mean
```

```python
words = []
pos = []
for el in df['Text']:
    words.extend(el.split())
for p in df['POS']:
    pos.extend(p.split())
```
